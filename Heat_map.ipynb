# Heatmap

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'Data/col_mean_demo.xlsx'
raw_data = pd.read_excel(file_path)

# Set the first column as row names and remove it from the dataframe
raw_data.set_index(raw_data.columns[0], inplace=True)

# Change column names
raw_data.columns = ["G1", "G2", "G3", "G4", "G5"]

# Row-wise min-max normalization
data = raw_data.apply(lambda x: (x - x.min()) / (x.max() - x.min()), axis=1)
```

```python
data
```

```python
# 1. Draw original color style heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data, 
            annot=False, fmt=".1f",
            cmap='vlag', cbar=True,
            linewidth=.5, 
            )
plt.show()
```

```python
# 2. Heatmap with clustering
sns.clustermap(
            data, 
            annot=False, fmt=".1f",
            row_cluster=True,
            cmap='vlag', cbar=True,
            linewidth=.5, 
            )
```
